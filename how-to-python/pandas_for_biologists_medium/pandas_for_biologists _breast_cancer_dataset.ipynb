### Read & Manipulate Data
The script below shows how to load the data set by using the pd.read_csv function from pandas. This functions reads .txt or .csv files directly from a folder on your machine and displays the data in a format that will remind you of a classic spreadsheets. We attribute the output of the function to a variable (breast_cancer_data) in order to use it along the script.

```python
# load the data setting the path_to_file/filename and the type of separator (sep)
# attribute the output of the function (our table) to a variable (breast_cancer_data)

breast_cancer_data = pd.read_csv('mydatasets/breast_cancer_diagnostic_data.csv', sep = ',')
breast_cancer_data
```

other parameters can be used inside the function to deal with different types of data <br>
for example, different separators, unwanted rows or columns, variable names, and other.

Note that we are simply checking the output of a given function here, not changing the data permanently. In order to do that, we need to attribute a new variable to the output of each operation. Alternatively, we can also use the parameter `inplace = True`, which tells pandas that we want to implement the action directly on the original table. To wrap up this first section, we can build a subset of the data (using some of the functions above) to work with a smaller table on the next section (`breast_cancer_data_subset`).

```python
# create a subtable containing all the rows and the first 7 columns 
breast_cancer_data_subset = breast_cancer_data.iloc[:, :8]

# we discard the columns containing patient id (we don't need here)
breast_cancer_data_subset.drop('id', axis = 1, inplace = True)

# we rename the columns with better/shorter names
# we do this by passing a new list of names
breast_cancer_data_subset.columns = ['diagnosis', 'radius', 'texture', 'perimeter','area', 'smoothness', 'compactness']

# tables can be saved using .to_csv function
# several parameters can be used to control the shape of the table
breast_cancer_data_subset.to_csv('mydatasets/my_new_table.txt', sep = '\t', index = False)
    
# and we have our new table
breast_cancer_data_subset
```

### Basic Operations

Two useful tools in pandas when starting to explore big data sets are the `pd.describe()` function, which returns a summary statistics for all numerical columns, and the `pd.corr()` function, which returns the correlation between all the columns in our data frame.

```python
breast_cancer_data_subset.describe()
```

```python
breast_cancer_data_subset.corr()
```

Note how much information can be retrieved from the data with just two simple lines of code! These functions use only columns containing numerical data and for that reason the diagnosis column is not shown. Other simple operations can be applied following a similar pattern. Here's a few more examples.

```python
# check the mean of a given column
breast_cancer_data_subset['radius'].mean()
```

```python
# check the standard deviation of a given column
breast_cancer_data_subset['radius'].std()
```

```python
# check the maximum or the minimum value of a given column
breast_cancer_data_subset['radius'].max(), breast_cancer_data_subset['radius'].min()
```

```python
# count number of elements of a given column
breast_cancer_data_subset['radius'].count()
```

```python
# sum all values in a given column
breast_cancer_data_subset['radius'].sum()
```

```python
# check the mean value of each column on the table
breast_cancer_data_subset.mean()
```

```python
# check the mean value of each line on the table
breast_cancer_data_subset.mean(axis = 1)
```

```python
# these operations can also be applied to slices of data at once
# e.g sum the radius, perimeter and area of the tumor of each patient

breast_cancer_data_subset[['radius','perimeter','area']].sum(axis = 1)
```

```python
# if we define this as a new parameter (dummy_col)
dummy_col = breast_cancer_data_subset[['radius','perimeter','area']].sum(axis = 1)

# we can add it to our table using the assign function
# sets the column name (string) and the variable that goes with it
breast_cancer_data_subset.assign(new_col = dummy_col)
```

```python
# to replace values by other values dynamically we can use replace function
# to replace 'M' and 'B' by the full word 'Malign' and 'Benign', respectively:

breast_cancer_data_subset['diagnosis'].replace(to_replace = ['M','B'], value = ['Malign','Benign'])
breast_cancer_data_subset
```

```python
# to answer the question "how many patients had a positive test result?"
# we use the value_counts() function: counts unique values in categorical variables

breast_cancer_data_subset['diagnosis'].value_counts() 
```

```python
# using the argument 'normalize' we get our answer in percentages

breast_cancer_data_subset['diagnosis'].value_counts(normalize =  True)
```

```python
# to sort our table according to a specific column we use sort_values()

breast_cancer_data_subset.sort_values(by = 'area', ascending = False)
```

Overall, sorting, counting, slicing and basic mathematical operations can be applied effortless. All functions typically contain optional parameters that can give us even more control over their outputs. detailed information concerning the role of each parameter can be found in the pandas documentation.

### Visualize Data

All the operations we made above are possible because pandas is built on top of `numpy`, the fundamental package for scientific computing. In the same way, `pandas` also incorporates part of the `matplotlib` library to create quick data visualizations. For this, we use the `plot()` function and set our variables ( `x` and `y` ) and the type of plot (kind) as parameters of that function.

```python
# to build a scatter plot with to variables
breast_cancer_data_subset.plot(kind = 'scatter', x = 'radius', y = 'texture');
```

```python
# by default pandas plots already look quite nice
# exploring the parameters we can create something even more fancy

breast_cancer_data_subset.plot(kind = 'scatter', x = 'radius', y = 'texture',  figsize = (6,5), alpha = 0.6, 
                               color = 'blue', s = 80, edgecolor = 'w', label = 'Malign', grid = False)
```

```python
# following the same rationale we can create a histogram for a given parameter 
breast_cancer_data_subset.plot(kind = 'hist', y = 'radius', bins = 10,
                              color = 'lightblue', edgecolor = 'black', width = 1.5);
```

```python
# we can also just select the column we want to plot
breast_cancer_data_subset['radius'].plot(kind = 'hist', edgecolor = 'black', width = 1.5)
```

```python
# and even plot mutiple histograms at once
breast_cancer_data_subset[['radius','texture']].plot(kind = 'hist', bins = 15, alpha = 0.8, width = 1.8,
                                                     color = ['lightblue','orange'], edgecolor = 'black');
```

```python
# we can also combine plot with the previous functions
# for instace, value_counts() with a barplot to visualize number of benign and malign cancers
breast_cancer_data_subset['diagnosis'].value_counts(normalize = True).plot(kind = 'bar', color = 'lightblue', 
                                                                fontsize = 12, width = 0.5, edgecolor = 'blue', rot = 0)
```

### Filter & Group

To conclude, let me introduce you data filtering using simple conditional arguments and the `groupby` function, arguably one of the best tools `pandas` has to offer. Conditionals follow a simple pattern, we define the condition we want to apply to a column or a row (e.g `>`, `<` or `==` than something) and we pass that condition in the same way we call for a column. Multiple conditions can be combined using the symbols`|` (meaning "or") and `&` ( meaning of "and"). Here's some examples.

```python
# lets say we want to filter patients that showed tumors with a area superior to 1000
our_condition = breast_cancer_data_subset['area'] > 1000
output = breast_cancer_data_subset[our_condition]
output
```

```python
# the output is a new table fullfilling that condition
# of course, we don't need to break into pieces, usually it's written

breast_cancer_data_subset[breast_cancer_data_subset['area'] > 1000]
```

```python
# now if we want to know from those tumors how many had a perimeter superior to 180?
# we combine both condition
breast_cancer_data_subset[(breast_cancer_data_subset['area'] > 1000) & (breast_cancer_data_subset['perimeter'] > 180)]    
```

```python
# if we want to know the mean tumor area of patients with a benign diagnosis
breast_cancer_data_subset[breast_cancer_data_subset['diagnosis'] == 'B']['area'].mean()
```

`groupby` function splits data into groups according to a categorical or numerical label in a given column. Then, mathematical operations can be applied to each one of those groups individually. In our example, the two labels in the diagnosis column (B or M) can be used for this purpose, allowing to investigate the data inside each condition.

```python
# here we use groupby to separate the data in two groups (benign and malign)
breast_cancer_data_subset.groupby('diagnosis')
```

```python
# the function itself has no output, some sort of function has to be applied
# lets say we want to know the median value of every parameter in each group

breast_cancer_data_subset.groupby('diagnosis').median()
```

```python
# we can apply the operations to specific columns

breast_cancer_data_subset.groupby('diagnosis')['radius'].median()
```

```python
# lets say we want to check correlation between parameters again 
# but this time inside each group (benign vs. malign)

breast_cancer_data_subset.groupby('diagnosis').corr()
```

```python
# this creates the same correlation maxtrix as we've seen before but for each group
# and we can still use seaborn to visualize the heatmap of that matrix

sns.heatmap(breast_cancer_data_subset.groupby('diagnosis').corr(), cmap = 'Reds', linecolor = 'black', linewidths = 1);
```
