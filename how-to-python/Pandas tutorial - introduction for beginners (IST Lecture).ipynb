**Some basic built-in `Pandas` functions**

```python
pokemon_data.columns              # check the headers
pokemon_data.shape                # check the size of your data
pokemon_data['Name']              # get Name column; each column is a Pandas Series (same as an array but with an index)
type(pokemon_data['Name'])        # check type; some functions might work only with specific types of data
pokemon_data.iloc[60]                   # check pokemon number 60 on the list
pokemon_data[['Name','Attack','Speed']] # get a subtable selecting mutiple columns (slice by columns)
pokemon_data.iloc[300:400]              # slice by rows 

# more tricky slicing; you can think as lists
# everything on the right side is related with columns and everything on the left side with rows
# by number
pokemon_data.iloc[:10,] # first 10 rows, all the columns
pokemon_data.iloc[:,4:8] # all the rows, column 5 to 9
pokemon_data.iloc[9:19,4:8] # arow 10:20, column 5 to 9

# by name
pokemon_data.loc[:,['Name','HP','Defense']]
pokemon_data.loc[:10,['Name','HP','Defense']];
```

## Statistics with `Scipy`
- SciPy contains additional routines needed in scientific work <br>
- computing integrals numerically, optimization, fitting equations and statistical tests
- Doc: https://docs.scipy.org/doc/scipy/reference/stats.html 

```python
# Load Main Table Again with our two new parameters
pokemon_data = pd.read_csv('data_examples/pokemon_data.csv')
pokemon_data['Overall'] = pokemon_data[['HP','Attack','Defense','Speed']].sum(axis = 1)
pokemon_data['Atk. Efficiency'] = pokemon_data['Attack'] * pokemon_data['Sp. Atk'];
pokemon_data.head()
```

**Let's check if the speed of first and sixth generation pokemons are significantly different?**

```python
# Let's create a list of pokemons for each generation
gen1 = pokemon_data[pokemon_data['Generation'] == 1]
gen6 = pokemon_data[pokemon_data['Generation'] == 6]

#gen2 = pokemon_data[pokemon_data['Generation'] == 2]
#gen3 = pokemon_data[pokemon_data['Generation'] == 3]
#gen4 = pokemon_data[pokemon_data['Generation'] == 4]
#gen5 = pokemon_data[pokemon_data['Generation'] == 5]
```

```python
# Are the mean speed of pokemons in generation 6 different from the ones in generation 1?
# let's plot boxplots ; make them pretty

plt.figure(figsize = (4,3), dpi = 120)
plt.boxplot([gen1['Speed'],gen6['Speed']],patch_artist = True,
            widths = 0.5, labels = ['Generation 1','Generation 6'])
plt.ylabel('Speed');

# we that there is a difference, but is it significant?
```

```python
#import scipy; statistical tests are in a sub-module called 'stats'

import scipy.stats

#apply a t-test
scipy.stats.ttest_ind(gen1['Speed'],gen6['Speed'], equal_var = False)
t_stats, pval = scipy.stats.ttest_ind(gen1['Speed'],gen6['Speed'], equal_var = False)
pval < 0.05

# extra tip: Seaborn (below)
# because changing colors and format for boxplots is not trivial!
```

## Fitting Models to Data `Scipy`

**Let's start simple with a linear regression**

```python
# plot pokemon_data['Attack'] vs. pokemon_data['Atk. Effiency']
plt.figure(figsize = (4,3), dpi = 150)
plt.plot(pokemon_data['Attack'],pokemon_data['Atk. Efficiency'], 'o', color = 'lightblue', alpha = 0.9, markeredgecolor = 'black')

# there's a package for linear regression in scipy.stats
# estimates the parameters from the linear regression
scipy.stats.linregress(pokemon_data['Attack'], pokemon_data['Atk. Efficiency'])
param = scipy.stats.linregress(pokemon_data['Attack'], pokemon_data['Atk. Efficiency'])
m,b,r_sq,pval,stderr = param

# let's visualize the regression
# we can define our new axis for the linear plot as
x_axis = np.linspace(0,200,500)
y_axis = m * x_axis + b

plt.plot(x_axis,y_axis, color = 'red', lw = 2, ls = '-')
'linear regression with a r-squared = {0:.3}'.format(r_sq)
```

**But we can also fit other 'custom' equations:**

```python
# we can define any other function with a lambda expression as we've done before at some point
# let's write an exponential function f

f = lambda x,a,b: a*(b**x)

scipy.optimize.curve_fit(f, pokemon_data['Attack'],pokemon_data['Atk. Efficiency'], p0 = [1,1])
param, cov = scipy.optimize.curve_fit(f, pokemon_data['Attack'],pokemon_data['Atk. Efficiency'], p0 = [1,1])

x_axis = np.linspace(0,200,100)
y_axis = f(x_axis,param[0],param[1])

plt.figure(figsize = (4,3), dpi = 150)
plt.plot(pokemon_data['Attack'],pokemon_data['Atk. Efficiency'], 'o', color = 'lightblue', alpha = 0.9, markeredgecolor = 'black')
plt.plot(x_axis,y_axis, '-r')

# r_squared estimation is not included
```

```python
# imagine we want to plot all files without even look at them!
# plot the directionality of all files!

plt.figure(figsize = (4,3), dpi = 120)

for file in files_of_interest:
    data = pd.read_csv(file, sep = ',')
    plt.hist(data['directionality'], bins = 30, alpha=0.7, edgecolor = 'black', label = 'Steroids');
```

```python
#change colors wen plotting

plt.figure(figsize = (4,3), dpi = 120)

for n, file in enumerate(files_of_interest): #check what the `enumerate` function is doing! it's cool !
    mycolors = ['red','green','purple']
    data = pd.read_csv(file, sep = ',')
    plt.hist(data['directionality'], color = mycolors[n],
             bins = 30, alpha=0.3, edgecolor = 'black', label = 'Steroids');
```

```python
# or we can quickly get info about a parameter from all distributions without open all the files
# example: get mean displacement, distance and directionality from the three samples

directionality_means = []
for file in files_of_interest:
    data = pd.read_csv(file, sep = ',')
    m = data['directionality'].mean()
    directionality_means.append(m)
    
distance_means = []
for file in files_of_interest:
    data = pd.read_csv(file, sep = ',')
    m = data['distance'].mean()
    distance_means.append(m)
    
displacement_means = []
for file in files_of_interest:
    data = pd.read_csv(file, sep = ',')
    m = data['displacement'].mean()
    displacement_means.append(m)

plt.figure(figsize = (4,3), dpi = 120)
sns.boxplot(data = pd.DataFrame([directionality_means,distance_means, displacement_means],
                                index = ['directionality','distance', 'displacement']).T);

# we can uptade the files, increase the number of file, etc , and we just need to run the code again .... 
# great to automatize boring stuff ... 
```
