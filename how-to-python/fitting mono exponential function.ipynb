**Import the Basics**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

**Import Data with Pandas**

```python
correlation_data = pd.read_csv('data_examples/Autocorrelation_data.txt', sep = '\t')
```

**Take a look how pretty the data looks like** <br>
Check the first 15 lines

```python
correlation_data.head(15)
```

**Organize Data**

```python
x_axis = correlation_data['Intervals']
y_axis = correlation_data['Average']
errors = correlation_data['Std']
```

**Plot Data with Shadow Bars**

```python
fig, ax = plt.subplots(figsize=(4, 3), dpi = 150)

plt.plot(x_axis,y_axis, '-', lw = 1.5,  color = 'blue', label = 'FtsZ Temp. Corr.')
plt.fill_between(x_axis,y_axis + errors, y_axis - errors, color = 'blue', alpha = 0.3)

#make the plot pretty
plt.xlabel('Time (frames)', fontsize=12)
plt.ylabel('Correlation Coefficient', fontsize=12)
plt.legend(loc = 0, frameon=False, fontsize = 8)
plt.tick_params(top=False, right=False)
#plt.xlim([-0.5,6.5]);
plt.ylim([0,1]);
plt.tick_params(direction = 'in', top=False, right=False, width = 1)
```

**Define the Desire Exponential Function to Fit the Data**

```python
def decay_fit(t_axis, a, b, k):
    return a * np.exp(-b*t_axis) + k
    
fitting_parameters, covariance = curve_fit(decay_fit, x_axis, y_axis, 
                                        bounds=((-np.inf, -np.inf, -np.inf), (1,np.inf, np.inf)))
a, b, k = fitting_parameters
x_fit = np.linspace(x_axis.values[0],x_axis.values[-1],100)

new_y_values = decay_fit(x_fit, *fitting_parameters)

#SAME PLOT AS BEFORE
fig, ax = plt.subplots(figsize=(4, 3), dpi = 150)
plt.plot(x_axis,y_axis, '-', lw = 1.5,  color = 'blue', label = 'FtsZ Temp. Corr.')
plt.fill_between(x_axis,y_axis + errors, y_axis - errors, color = 'blue', alpha = 0.3)

# AND THE PLOT WITH THE FITTING EQUATION
plt.plot(x_fit, new_y_values, 'r')

#make the plot pretty again
plt.xlabel('Time (frames)', fontsize=12)
plt.ylabel('Correlation Coefficient', fontsize=12)
plt.legend(loc = 0, frameon=False, fontsize = 8)
plt.tick_params(top=False, right=False)
#plt.xlim([-0.5,6.5]);
plt.ylim([0,1]);
plt.tick_params(direction = 'in', top=False, right=False, width = 1)
```

**Extract Half time**

```python
# using the fitting parameters
a, b, k = fitting_parameters

# halftime is given by
half_time = -np.log(abs((a-k)/(2*a)))/b

# cool displaying of results:
'half_time of the decay is {:.3} frames'.format(half_time)
```
