```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### import data

```python
sample1 = pd.read_csv('data_examples/gamma_file1.csv').X1.dropna()
sample2 = pd.read_csv('data_examples/gamma_file2.csv').X1.dropna()

# obviously this has to be adpated to the type of data you want to use ...
```

### How to Fit Gamma Function to a Distribution

```python
from scipy.stats import rayleigh, gamma

def GammaFit(sample):
    '''returns x_axis and y_axis of a fitting curve using Gamma Function'''
    
    gamma_param = gamma.fit(sample, loc = 0)

    counts, bins = np.histogram(sample, density = True) # normalizes data (area under the curve = 1)
    bins_center = (bins[:-1] + np.diff(bins) / 2) #calculate bin center to use as y data to fit
    x_axis = np.linspace(0, bins[-1]) #creates an arbitrary x_axis to plot the fit

    gamma_fit_pdf = gamma.pdf(x_axis, *gamma_param) #uses gamma paramters to create the y-axis 
    
    return x_axis, gamma_fit_pdf, gamma.mean(*gamma_param)
```

```python
#plot pretty histogram
fig, ax = plt.subplots(figsize = (5,4), dpi = 120)
bin_number = 20

plt.hist(sample1, bins = bin_number, density = True, color = 'steelblue', alpha = 0.3, edgecolor = 'w', label = 'sample1')
plt.plot(GammaFit(sample1)[0], GammaFit(sample1)[1], lw = 2, color = 'steelblue', label = 'Gamma Fit sample1 ')

plt.hist(sample2, bins = bin_number, density = True, color = 'red', alpha = 0.3, edgecolor = 'w', label = 'sample2')
plt.plot(GammaFit(sample2)[0], GammaFit(sample2)[1],  lw = 2, color = 'red', label = 'Gamma Fit sample2')

plt.xlabel('some units'); plt.ylabel('PDF')
plt.legend(fontsize = 9, frameon = False)

#save figures
plt.savefig('plot_examples/Gammafit.png', transparent=False, bbox_inches = 'tight')

# Print means
'sample1 mean = {:4.3}; sample2 mean = {:4.3}'.format(GammaFit(sample1)[2], GammaFit(sample2)[2])
```

### How to fit a Rayleigh function? Same thing ... 

```python
def RayleighFit(sample):
    '''returns x_axis and y_axis of a fitting curve using Gamma Function'''
    
    ray_param = rayleigh.fit(sample, loc = 0)

    counts, bins = np.histogram(sample, density = True) # normalizes data (area under the curve = 1)
    bins_center = (bins[:-1] + np.diff(bins) / 2) #calculate bin center to use as y data to fit
    x_axis = np.linspace(0, bins[-1]) #creates an arbitrary x_axis to plot the fit

    ray_fit_pdf = rayleigh.pdf(x_axis, *ray_param) #uses gamma paramters to create the y-axis 
    
    return x_axis, ray_fit_pdf, rayleigh.mean(*ray_param)
```

```python
fig, ax = plt.subplots(figsize = (5,4), dpi = 120)
bin_number = 20

plt.hist(sample1, bins = bin_number, density = True, color = 'steelblue', alpha = 0.3, edgecolor = 'w', label = 'sample1')
plt.plot(RayleighFit(sample1)[0], RayleighFit(sample1)[1], lw = 2, color = 'steelblue', label = 'Rayleigh Fit sample1 ')

plt.hist(sample2, bins = bin_number, density = True, color = 'red', alpha = 0.3, edgecolor = 'w', label = 'sample2')
plt.plot(RayleighFit(sample2)[0], RayleighFit(sample2)[1],  lw = 2, color = 'red', label = 'Rayleigh Fit sample2')

plt.xlabel('units'); plt.ylabel('PDF')
plt.legend(fontsize = 9, frameon = False)

'sample1 mean = {:4.3}; sample2 mean = {:4.3}'.format(RayleighFit(sample1)[2], RayleighFit(sample2)[2])
```
